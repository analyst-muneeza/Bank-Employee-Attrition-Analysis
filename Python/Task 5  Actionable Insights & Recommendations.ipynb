# Task 5 – Actionable Insights & Recommendations

# Objective

Identify key attrition drivers, define a high-risk employee profile, and recommend data-driven retention strategies based on the analysis performed.

# Top Attrition Drivers

# Insight 1: Short Employee Tenure

- Finding

Employees in the lower tenure bands showed higher attrition compared to long-tenured employees.

- Business Impact

New employees may leave before becoming fully productive, increasing recruitment and training costs.

# Insight 2: Department-Specific Attrition

- Finding

Certain departments recorded higher attrition rates than others.

- Business Impact

Departmental attrition can create staffing shortages and impact customer service quality.

# Insight 3: Low Performance Ratings

- Finding

Employees with lower Quarterly Ratings showed a higher tendency to leave.

- Business Impact

Performance-related challenges may contribute to disengagement and turnover.

# Insight 4: Limited Income Growth

- Finding

Employees who experienced little or no income increase were more likely to leave.

- Business Impact

Lack of compensation growth can reduce employee satisfaction and retention.

# High-Risk Employee Profile

Based on the analysis, a typical employee likely to leave may have the following characteristics:

| Factor        | High-Risk Profile                 |
| ------------- | --------------------------------- |
| Tenure        | Less than 12 months               |
| Department    | Department with highest attrition |
| Rating Trend  | No improvement in ratings         |
| Income Growth | No recent salary increase         |
| Grade         | Lower grade employees             |
| Overtime      | Frequent overtime work            |

- Conclusion

Employees matching multiple risk factors should be monitored closely through HR retention programs.


# Retention Strategies

| Strategy                            | Supporting Finding                                           | Recommended Action                                                       |
| ----------------------------------- | ------------------------------------------------------------ | ------------------------------------------------------------------------ |
| Improve Early Employee Experience   | Highest attrition observed among employees with short tenure | Introduce structured onboarding and mentorship programs                  |
| Performance Development Programs    | Employees with lower ratings are more likely to leave        | Provide regular training, coaching, and career development opportunities |
| Transparent Compensation Growth     | Employees without income growth show higher attrition        | Link salary increments more closely to performance outcomes              |
| Department-Specific Retention Plans | Certain departments show higher attrition                    | Conduct department-level engagement surveys and targeted interventions   |
| Work-Life Balance Initiatives       | Overtime may contribute to employee stress                   | Promote flexible work schedules and workload management                  |


# Top 5 Findings

- Employee attrition is concentrated among employees with shorter tenure.
- Attrition rates vary significantly across departments.
- Employees with lower performance ratings are more likely to leave.
- Limited income growth may contribute to employee turnover.
- Workload and overtime can influence employee retention.

# Final Summary

The analysis identified tenure, department, performance ratings, income growth, and workload as important factors associated with employee attrition. Employees with short tenure, limited salary progression, and weaker performance trends showed a greater likelihood of leaving the organization. Implementing targeted onboarding, career development, compensation, and employee engagement strategies can help reduce attrition and improve workforce stability.
